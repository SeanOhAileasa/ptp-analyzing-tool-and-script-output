### Functions in Scripting
*** 

Functions are small blocks of code that are embedded within our script and that we can call on to perform an action at any time. 

Can think of a function like a script within a script - it's a small block of code that we'll be calling often multiple times throughout a script, because usually a function represents an action that we want to carry out over and over again.

A good example of a function might be an encryption function in a security piece of code and we might want to encrypt many different values, but we don't want to have to write out all the code associated with performing an encryption every time we want to go and encrypt a value, so instead, we'll write out that code once, in a general form within a function and then, when we want to encrypt a value, we'll pass that value as an argument into the function and then all of that code is executed to encrypt our value and returned to us through that function and we only ever have to write the encryption code once and then we just pass it values from then on. 

This is very useful for reducing the overall amount of code that's within your script, so your maintainability is much higher because if you need to change something, for example, functionality of a function, you don't need to change it in ten different places, you only change it in one place within your function.

The general format of a function is as follows: 

Start by defining the function name and you want to give a name that's descriptive of what the function is doing, so using our encryption function example, we might have it called something like encrypt or encrypt value. 

The name format depends on the language you're using. If you're using something like Java, then you want to use camelCase, where the first letter of the first word is lowercase, but the first letter of all subsequent words are uppercase.

If you're using Python, then you want to use Snake case, where we put an underscore between each name.

Then we have round parentheses, these round parentheses, following the function name, represent where the arguments will be passed to the function. In the example on the screen, there is no value between the round parentheses, meaning this function doesn't take a value and that's OK, they don't have to take a value. If it took a value, you might see something like string my_string in there, telling the function it takes in a string value and does something with that string value. 

Depending on the language, once again, you may or may not have visibility identifiers, so you might have something like public or private in front of the function to define the visibility and you might also have a return type, for example, in Java, if we're returning a string, then we would have the keyword string in front of our function name to tell Java this function returns a string. 

After the round parentheses, we have an open curly brace to signify the start of the actual commands within a function, so once again, with our encryption example, this is where all of the code to perform an encryption would happen between the open curly brace and the following closed curly brace, which signifies the end of our function.

So once again, functions are great for reducing the amount of duplicate code in your scripts, and they're ideal for performing repetitive tasks. 

A good rule of thumb is if you find yourself writing the same piece of code twice just with different values being applied, then that's an ideal scenario to rework your code into a function.

When we look at functions, we have to look at the scope of our variables, what this means is that we'll have many different variables defined within a script - we might have certain set up variables defined right at the beginning of the script and then we might have variables defined within our function itself and the accessibility or visibility of these variables, by the script as a whole, is very different, so any variables defined in the main part of the script, we would define as global variables, and they're accessible to any sub part of the script, such as a function, so if you have a function, it can access variables that are defined outside of itself in the main part of the script. 

However, we can also define local variables within a function, and those variables are only accessible from within the function they're defined inside of, so if you define a variable inside of a function, the main part of the script has no access or awareness of that variable, so you need to consider when defining a variable, does this variable need to be accessed by the entire script or just by the function that is currently being created?

Then we have function arguments. These are the values that are passed to a function in order for it to operate on. There is no limit to how many arguments can be passed to a function - can pass no arguments or you can pass any number of arguments and then these arguments will be available to the function to work on. 

Oftentimes, we want our functions to return some sort of value to us. Not all functions return a value, for example, maybe you have a function that is solely designed to print something to the command line and then exit - in that case, it doesn't need to return a value but going back to our encryption example, the encryption function is going to take in a plaintext value and you want it to return an encrypted text value. In that case, we would use the return command, which would be used to return a value from a function. 

The syntax is as follows: we use the word return and then we pass the value that we want to return. This could be a variable or an explicitly defined string or integer, whatever it is, whatever it's after, the return gets returned outside of the function, so oftentimes what we'll do is we'll have a variable and we'll say this variable is equal to the result of this function and then it's assigned whatever is returned from that function, so the actual value that's returned can be anything specified in the script.
