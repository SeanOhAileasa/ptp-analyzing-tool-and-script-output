### Classes in Scripting
*** 

Classes in programming are a method that we can use to bundle similar functionality and types of data together. 

Classes can be used in a wide array of situations. If we have a type of object that we want to create over and over again with just slightly different parameters, that's a prime opportunity for a class, for example, let's say in our program, we want to define many different types of cars - well, then we could define a car class that would have all the attributes of a car like color, make, model, and price, and then we could create new instances of that car class with the fields filled out appropriately.

Classes also give us a way to maintain the states of these objects - let's say we gave our car a paint job - could update the state of the color from red to blue and then it's permanently altered in that object.

Not all programming languages support the concept of a class. 

The class structure tends to be found more in object-oriented programming languages. 

Python supports classes because it is an object-oriented programming language. One of the nice things about Python compared to other programming languages is that when it comes to defining your classes in terms of the structure, the syntax and semantics associated with that definition tends to be much less verbose and much easier to read.

Once you have a class defined, you can start to use the class inheritance, where base classes can inherit features from parent classes and derived classes can overwrite any methods that are found in other classes.

In C++, when we're defining a class, we'll often have fields and variables defined within a class, as well as functions or methods defined within a class. 

When it comes to these, we can assign visibility identifiers such as public or private, and the identifier we give it depends on how we want the field to be seen by the rest of the package.

If we want it to be available only to the class in which it's defined, we can make something private, but if we want it to be freely available to any part of the code, we would define it as public and this refers to the variables as well as the functions.

Functions can also be virtual, which serves as a template for any derived classes that are implementing a function in order to maintain a certain structure - it's a form of polymorphism where the actual implementation of the function is defined at runtime.

Defineing a class - the definition statement for a class in Python - start by using the keyword class and that tells the Python interpreter that we're about to define the name for a class. 

Then, following that, we have the name - in this case called ClassName - generally, you want it to be descriptive of what the class is doing. 

Finish the declaration with a colon and that tells Python, here's the end of our class declaration, now we're going to start getting into the definition of the class itself. 

Within the class, we can have a variety of things.

We could have definitions of variables, methods or general statements, and we can have any number of these statements and these statements are all indented because then the Python interpreter knows everything following that colon on the next lines that is indented is part of the class.

These statements could be something as simple as X = 1, or more complicated terms of functionality.

Could have a class called MyClass, once again defined with the keyword class, the name MyClass and a colon and then in the indentation we have a comment which tells us, ``"""Example class"""`` and then we have a statement saying i = abcde. 

Note that these are both indented so that they are part of MyClass. 

Then we could define a method within MyClass and in this case we use ``def`` to tell the interpreter that we're starting to define a method - it's called ``f`` and it takes the parameter of (self) and then, that method has an indented statement, which is part of the method, which is part of the class, to return the string ``'hello world'`` and then all of these are available when you instantiate the class MyClass, so if you use an instance of MyClass, you could call the ``f`` method and it would return you the string ``'hello world'``.

Could get an instance of MyClass and retrieve the ``i`` variable which has the value ``abcde`` in it.

The advantages of classes are:

- Have a set data structure that's consistent and predictable


- Can copy instances of the classes between scripts or share them to reduce the duplication of code 


- Keep data and functions portable
